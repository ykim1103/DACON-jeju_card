# 'CSTMR_CNT','CNT','AMT'

## 표준화

```python
# CSTMR_CNT ,CNT,AMT열을 가져와서 따로 데이터프레임을 만든 후 표준화하기

data2 = data[['CSTMR_CNT','CNT','AMT']]
data2


data_standadized = ss.zscore(data2)
```

```python
data_standadized_df = pd.DataFrame(data_standadized)
data_standadized_df
```

## 정규화


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler =  MinMaxScaler()
```

```python
scaler_data2 = scaler.fit_transform(data_standadized_df)
```

```python
scaler_data2_df = pd.DataFrame(scaler_data2,columns=['CSTMR_CNT','CNT','AMT'])
scaler_data2_df
```

## 상관관계

```python
corr =  scaler_data2_df.corr(method='pearson')
```

```python
corr
```

# AMT.sum(), holiday

```python
data3=data[['REG_YYMM','AMT','Holiday']]
data3
```

```python
data3=data3.groupby(['REG_YYMM','Holiday'],as_index=False).sum()
data3['days']=[31,28,31,30,31,30,31,31,30,31,30,31,31,29,31]
data3
```

```python
data3['rate_holiday']= (data3['Holiday']/data3['days']).round(2)
data3
```

```python
data4=data3[['AMT','rate_holiday']]
data4
```

## 표준화

```python
data4_standadized_AMT = ss.zscore(data4['AMT'])
data4['AMT']=data4_standadized_AMT
data4
```

## 정규화

```python
from sklearn.preprocessing import MinMaxScaler
amt_scaler =  MinMaxScaler()
scaler_data4 = amt_scaler.fit_transform(data4)
```

```python
scaler_data4=pd.DataFrame(scaler_data4,columns=['AMT','rate_holiday'])
```

```python
scaler_data4
```

## 상관관계

```python
corr =  scaler_data4.corr(method='pearson')
corr
```
