```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
data = pd.read_csv('data_preprocessing_1.csv')
```

```python
data.head()
```

```python
### Holiday 삭제
data.drop('Holiday', axis = 1, inplace = True)
data.head()
```

```python
# 년, 월 쪼개는 함수
def grap_year(data):
    data = str(data)
    return int(data[:4])

def grap_month(data):
    data = str(data)
    return int(data[4:])
```

```python
# 세종시 해결  (CARD_CCG_NM 값이 nan이다)

data_sj =data[data['CARD_SIDO_NM']=='세종']
data_sj.fillna('세종시',inplace=True)



# 년,월 쪼개기
data_sj['year'] = data_sj['REG_YYMM'].apply(lambda x: grap_year(x))
data_sj['month'] = data_sj['REG_YYMM'].apply(lambda x: grap_month(x))
data_sj = data_sj.drop(['REG_YYMM'], axis=1)
data_sj.head()


# 세종시 데이터프레임 정리
columns = ['year', 'month', 'CARD_SIDO_NM', 'CARD_CCG_NM', 'STD_CLSS_NM', 'SEX', 'region_diff2',
       'AGE', 'CSTMR_CNT', 'CNT', 'AMT']


data_sj = data_sj[columns]
data_sj.head()

data_sj
```

```python
data_sj = data_sj.groupby(['year','month', 'CARD_SIDO_NM', 'CARD_CCG_NM',  'STD_CLSS_NM', 'SEX', 'region_diff2', 'AGE']).sum()
data_sj = data_sj.reset_index(drop = False)
data_sj
```

```python
# REG_YYMM -> year, month
data['year'] = data['REG_YYMM'].apply(lambda x: grap_year(x))
data['month'] = data['REG_YYMM'].apply(lambda x: grap_month(x))
data = data.drop(['REG_YYMM'], axis=1)
data.head()
```

```python
# target인 AMT을 가장 뒤로
columns = ['year', 'month', 'CARD_SIDO_NM', 'CARD_CCG_NM', 'STD_CLSS_NM', 'SEX', 'region_diff2',
       'AGE', 'CSTMR_CNT', 'CNT', 'AMT']

data_final1 = data[columns]
data_final1.head()
```

```python
data_final1[data_final1['CARD_SIDO_NM']=='세종']
```

```python
# 모든 카테고리를 기준으로 groupby & FLC 제외

df_groupby = data.groupby(['year','month', 'CARD_SIDO_NM', 'CARD_CCG_NM',  'STD_CLSS_NM', 'SEX', 'region_diff2', 'AGE']).sum().iloc[:, 1:]
df_groupby = df_groupby.reset_index(drop = False)
df_groupby.head()
```

```python
df_groupby[df_groupby['CARD_SIDO_NM']=='세종']
```

```python
final_data = pd.concat([df_groupby,data_sj])
```
